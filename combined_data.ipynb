```python
# Import Dependencies and Setup
import pandas as pd
import csv
import os

# Files to Load
jul_2013_path = os.path.join(os.getcwd(), "Raw Data", "2013-07 - Citi Bike trip data.csv")
dec_2013_path = os.path.join(os.getcwd(), "Raw Data", "2013-12 - Citi Bike trip data.csv")
```

```python
#Print path
jul_2013_path
```

```python
# Read Files and Store Into Pandas Data Frame
jul_2013_df = pd.read_csv(jul_2013_path)
dec_2013_df = pd.read_csv(dec_2013_path)

#Display dataframe
jul_2013_df.head()
```

```python
## Combine the data into a single dataset
cleaned_2013_df = pd.concat([jul_2013_df, dec_2013_df], axis = 0)
cleaned_2013_df.head()
```

```python
#Checking for value types
cleaned_2013_df.dtypes
```

```python
# Making gender column a string
cleaned_2013_df['gender'] = cleaned_2013_df['gender'].astype(str)
```

```python
# Replace the numbers in the gender column with strings
cleaned_2013_df['gender'] = cleaned_2013_df['gender'].replace('1', 'Male')
cleaned_2013_df['gender'] = cleaned_2013_df['gender'].replace('2', 'Female')
cleaned_2013_df['gender'] = cleaned_2013_df['gender'].replace('0', 'Unknown')
cleaned_2013_df.head()
```

```python
#Finding empty values
cleaned_2013_df.isnull().values.any()
```

```python
#Convert the dataframe to a csv
cleaned_2013_df.to_csv(r'c:\\Users\\vaab2\\OneDrive\\Desktop\\Excercises in class\\Homeworks\\18-Tableau\\Resources\\cleaned_2013_data.csv', index = False)
```

```python
#Load paths
mar_2018_path = os.path.join(os.getcwd(), "Raw Data", "201803-citibike-tripdata.csv")
sep_2018_path = os.path.join(os.getcwd(), "Raw Data", "201809-citibike-tripdata.csv")
dec_2018_path = os.path.join(os.getcwd(), "Raw Data", "201812-citibike-tripdata.csv")
```

```python
# Read Files and Store Into Pandas Data Frame
mar_2018_df = pd.read_csv(mar_2018_path)
sep_2018_df = pd.read_csv(sep_2018_path)
dec_2018_df = pd.read_csv(dec_2018_path)

#Display df
mar_2018_df.head()
```

```python
## Combine the data into a single dataset
cleaned_2018_df = pd.concat([mar_2018_df, sep_2018_df, dec_2018_df], axis = 0)
cleaned_2018_df.head()
```

```python
#Checking for value types
cleaned_2018_df.dtypes
```

```python
# Making gender column a string
cleaned_2018_df['gender'] = cleaned_2018_df['gender'].astype(str)
# Replace the numbers in the gender column with strings
cleaned_2018_df['gender'] = cleaned_2018_df['gender'].replace('1', 'Male')
cleaned_2018_df['gender'] = cleaned_2018_df['gender'].replace('2', 'Female')
cleaned_2018_df['gender'] = cleaned_2018_df['gender'].replace('0', 'Unknown')
cleaned_2018_df.head()
```

```python
#Finding empty values
cleaned_2018_df.isnull().values.any()
```

```python
#Replacing blank values with a null
import numpy as np
#np.where(combined_2018_df.applymap(lambda x: x == ''))
cleaned_2018_df = cleaned_2018_df.replace(' ', np.nan)
#cleaned_2018_df.loc[cleaned_2018_df['birth year'] == ' ', 'birth year'] = 0
```

```python
cleaned_2018_df.head()
```

```python
#Find the riders age and creating an age column
current_year = 2021

birth_year_2018 = cleaned_2018_df['birth year']
age = current_year - (birth_year_2018)
cleaned_2018_df['age'] = age
#Display df
cleaned_2018_df.head()
```
